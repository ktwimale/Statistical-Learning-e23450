# Hypothesis Testing


Hypothesis testing operates on a philosophy similar to "innocent until proven guilty." We start by assuming a baseline state of no effect and see if our data provides enough evidence to reject that assumption.

* **The Null Hypothesis ($H_0$):** The status quo. It states that there is no effect, no difference, or no relationship. For example, $\mu_1 = \mu_2$ (the means of two groups are equal).
* **The Alternative Hypothesis ($H_a$ or $H_1$):** The claim you are actually trying to prove. It states that there *is* an effect or difference (e.g., $\mu_1 \neq \mu_2$).



## Mathematical Machinary

To evaluate the hypotheses, we compress our sample data into a single metric and compare it to what we would theoretically expect under the null hypothesis.

* **Test Statistic:** A standardized value calculated from your sample data (like a $z$-score, $t$-statistic, or $F$-statistic). The general formula often looks like:

$$\text{Test Statistic} = \frac{\text{Observed Value} - \text{Hypothesized Value}}{\text{Standard Error}}$$


* **The $p$-value:** The probability of obtaining a test statistic at least as extreme as the one observed, *assuming the null hypothesis is completely true*.
$$p\text{-value} = P(\text{Test Statistic} \geq \text{Observed} \mid H_0)$$
Here $P$ is the probability model (or probability distribution) of the specific test statistic, conditioned on the assumption that the null hypothesis ($H_0$) is true




## The Decision Rule & Errors

Before looking at the data, we set a threshold for risk called the **significance level ($\alpha$)**, typically set at **5%** (0.05).

* If $p\text{-value} \leq \alpha$: We **reject $H_0$**. The data is too unusual to have happened by chance alone. We say the result is "statistically significant."
* If $p\text{-value} > \alpha$: We **fail to reject $H_0$**. We do *not* "accept" the null; we simply state that we lack sufficient evidence to disprove it.

## The Standard Roadmap

Every hypothesis test follows these exact five steps:

1. **State the hypotheses** ($H_0$ and $H_a$).
2. **Formulate an analysis plan** (choose the right test—like a t-test, Chi-Square, or ANOVA—based on your data type and assumptions).
3. **Analyze sample data** (calculate the test statistic and $p$-value).
4. **Apply the decision rule** (compare the $p$-value to $\alpha$).
5. **Interpret the results** in the context of your original real-world problem.

## A/B Testing

A/B testing is simply the tech industry's and business world's name for a two-sample hypothesis test.

**A/B testing** is simply the tech industry’s and business world's name for a **two-sample hypothesis test**.

It is a controlled experiment where you split your audience into two groups to compare two versions of something (Version A and Version B) to see which one performs better based on a specific metric.

While academic statisticians talk about "treatment vs. control groups," data scientists and product managers talk about "A/B testing."

---

**Example**

Imagine you run an e-commerce website, and you want to know if changing your "Buy Now" button from **Blue** to **Green** will increase sales.

* **Group A (The Control):** $50\%$ of your website visitors see the original **Blue** button.
* **Group B (The Variation):** The other $50\%$ of visitors see the new **Green** button.

You let the experiment run for a few weeks, track how many people click the button in each group, and then use your hypothesis testing machinery to analyze the results.

## Introduction: The Multivariate i.i.d. Assumption

In rigorous statistical modeling, a dataset consisting of $m$ numerical features and $n$ observations is not merely a static matrix, but a collection of empirical observations generated by an underlying stochastic process. Formally, we define each row vector $\mathbf{x}_i \in \mathbb{R}^m$ (for $i = 1, 2, \dots, n$) as an observed **realization** of a sequence of $m$-dimensional multivariate random variables, denoted by $\{\mathbf{X}_i\}_{i=1}^n$.

To justify the application of classical multivariate statistical inference, we impose the foundational assumption that this sequence is **independent and identically distributed (i.i.d.)**. Mathematically, this implies that the joint Cumulative Distribution Function (CDF) of the entire sample factors perfectly into the product of identical marginal multivariate CDFs, expressed as:

$$F_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}(\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n) = \prod_{i=1}^{n} F_{\mathbf{X}}(\mathbf{x}_i) \quad \forall \mathbf{x}_i \in \mathbb{R}^m$$

Furthermore, assuming the multivariate random variables are continuous and the joint distribution is absolutely continuous, this system possesses a Radon-Nikodym derivative. This allows us to equivalently express the i.i.d. condition via the joint Probability Density Function (PDF), denoted by $f_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}$:

$$f_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}(\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n) = \prod_{i=1}^{n} f_{\mathbf{X}}(\mathbf{x}_i) \quad \forall \mathbf{x}_i \in \mathbb{R}^m$$

where $f_{\mathbf{X}}(\mathbf{x}_i)$ represents the identical marginal PDF evaluated at the $i$-th realization.

Under this joint-density framework, the **independence** condition dictates that the conditional density of any realization $\mathbf{X}_i$ given $\mathbf{X}_j$ ($i \neq j$) collapses to its marginal density ($f_{\mathbf{X}_i \mid \mathbf{X}_j} = f_{\mathbf{X}_i}$), ensuring that the row sequence possesses no serial or spatial memory. Concurrently, the **identically distributed** condition asserts that every realization originates from the exact same data-generating process governed by this time-invariant probability law, thereby sharing an identical, constant population mean vector $\boldsymbol{\mu} \in \mathbb{R}^m$ and a static variance-covariance matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{m \times m}$. When the sample size is massive relative to the feature space ($n \gg m$), these theoretical properties ensure that our sample estimators asymptotically converge to their true population parameters via the Law of Large Numbers, establishing the baseline validity required for subsequent statistical modeling.

## Empirical Verification via Weak i.i.d. Testing

While strict independence and identical distribution require all higher-order probability moments to be invariant and uncoupled across realizations, testing these infinite-dimensional criteria empirically is practically impossible. Therefore, in classical data analysis, we relax our requirements to **second-order (weak) conditions**.

To validate that our data satisfies the weak i.i.d. assumption, we partition our $n$ rows into $g$ distinct, consecutive, non-overlapping blocks (or chunks) of size $n_j$ (where $\sum_{j=1}^g n_j = n$). Because $n \gg m$, each individual chunk remains statistically large enough to yield highly precise sample estimators. We then verify three necessary conditions concerning the first two joint moments:



### Condition 1: Constant Joint Mean Vector (First Moment Homogeneity)



The expected value vector of the multivariate random variable must remain invariant across the sequence of realizations. The center of the distribution cannot drift or exhibit structural shifts over the row index.

* **Statistical Formulation:** Let $\bar{\mathbf{x}}_j \in \mathbb{R}^m$ be the empirical sample mean vector calculated for chunk $j$. We test the null hypothesis that all chunk population means are equivalent to a single global mean vector $\boldsymbol{\mu}$:

$$H_0: \mathbb{E}[\mathbf{X}_{i \in \text{chunk } 1}] = \mathbb{E}[\mathbf{X}_{i \in \text{chunk } 2}] = \dots = \mathbb{E}[\mathbf{X}_{i \in \text{chunk } g}] = \boldsymbol{\mu}$$


* **Verification Methodology:** This condition is evaluated using **Multivariate Analysis of Variance (MANOVA)** via Wilks' Lambda ($\Lambda$). MANOVA determines if the variance *between* the chunk mean vectors is significantly larger than the variance *within* the chunks.



### Condition 2: Constant Covariance Matrix (Second Moment Homogeneity)

The dispersion profile of the multivariate random variable and the linear relationships between its $m$ dimensions must be stable across the sample sequence. The data must not exhibit heteroscedasticity or structural breaks in its correlation fabric.

* **Statistical Formulation:** Let $\mathbf{S}_j \in \mathbb{R}^{m \times m}$ be the sample variance-covariance matrix evaluated within chunk $j$. We test the null hypothesis that the true population covariance structure is identical across all chunks:

$$H_0: \boldsymbol{\Sigma}_1 = \boldsymbol{\Sigma}_2 = \dots = \boldsymbol{\Sigma}_g = \boldsymbol{\Sigma}$$


* **Verification Methodology:** This structural equivalence is evaluated using **Box's M-test**, which relies on a likelihood-ratio test statistic comparing the determinants of the individual chunk covariance matrices ($|\mathbf{S}_j|$) against the determinant of the pooled sample covariance matrix ($|\mathbf{S}_p|$).


### Condition 3: Zero Cross-Covariance Between Realizations (Linear Independence)

Distinct rows must not carry memory or linearly predict subsequent rows. Any pair of realizations separated by a row distance (lag) $k$ must yield an expected cross-covariance of zero, effectively ruling out serial autocorrelation.

* **Statistical Formulation:** For any two non-identical row indices $i$ and $j$, the cross-covariance matrix must vanish:

$$H_0: \text{Cov}(\mathbf{X}_i, \mathbf{X}_j) = \mathbb{E}\left[ (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_j - \boldsymbol{\mu})^T \right] = \mathbf{0}_{m \times m} \quad \forall i \neq j$$



In a sequential framework, this implies that at any lag $k \neq 0$, the multivariate autocovariance matrix $\boldsymbol{\Gamma}(k) = \mathbf{0}_{m \times m}$.
* **Verification Methodology:** This condition is evaluated dynamically across the entire dataset using the **Multivariate Ljung-Box Test** (or Hosking's Portmanteau Test). It calculates the sample cross-correlation matrices $\mathbf{R}(k)$ up to a designated maximum lag $H$, testing if the collective serial correlation is statistically indistinguishable from a white-noise process.

# Principle Component Analysis

Consider an observation that is characterized by a $m$-dimensional random variable
\begin{align}
X&\triangleq \begin{bmatrix} X_{1} & X_{2} & \cdots & X_{m}
\end{bmatrix}^T.
\end{align}



Let
\begin{align}
\mathbb{E}(XX^T)=U\Sigma V^*
\end{align}
where $(U,\Sigma,V^*)$ is the SVD of $\mathbb{E}(XX^T)$.

Define $Z\triangleq U^*X$ and $F\triangleq V^*X$ then
\begin{align}
\mathbb{E}(ZF^T)=\Sigma
\end{align}
\begin{align}
XF^T=U\Sigma
\end{align}

Thus
\begin{align}
\mathbb{E}(FF^T)=\mathbb{E}(V^*XX^TV)=V^*\mathbb{E}(XX^T)V=V^*U\Sigma=\Sigma
\end{align}

# Factor Analysis

Let $Z$ be $m$-dimensional continuous or discrete valued random variable with zero mean and covariance $E(ZZ^T)=I_{m\times m}$. We seek to find $\Lambda_{mk}$ and a reduced $k$-dimensional random variable $F$ such that
\begin{align*}
Z=\Lambda F+\epsilon,
\end{align*}
where $\epsilon$ is a $m$-dimensional random variable with zero mean, $E(\epsilon \epsilon^T)=\Sigma_{\epsilon}$, and $E(\epsilon F^T)=0$.  We additionally impose the condition that $E(FF^T)=I_{k\times k}$.

The condition $E(\epsilon F^T)=0$ implies that
\begin{align*}
E(ZF^T)=\Lambda.
\end{align*}
Also
\begin{align*}
E(ZZ^T)=\Lambda\Lambda^T+E(\epsilon\epsilon^T)\\
E(z_iz_j)=\sum_{r=1}^k\lambda_{ir}\lambda_{jr}+E(\epsilon_{i}\epsilon_j)
\end{align*}
Thus in particular
\begin{align*}
E(z_i^2)=\sum_{r=1}^k\lambda_{ir}^2+E(\epsilon_{i}^2)=1,
\end{align*}
where the last inequality occurs due to the fact that $Z$ is normalized.

Let
\begin{align*}
h_i^2&\triangleq \sum_{r=1}^k\lambda_{ir}^2,\\
\varphi_i^2&\triangleq 1-\sum_{r=1}^k\lambda_{ir}^2=1-h_i^2=E(\epsilon_{i}^2).
\end{align*}
The parameter $h_i^2$ is called the \emph{communality} of $z_i$ and the parameter $\varphi_i^2=E(\epsilon_{i}^2)$ is called the \emph{uniqueness} of $z_i$.  

Since $E(z_if_j)=\lambda_{ij}$ we see that if $\varphi_i^2\approx 1$ then it indicates that $z_i$ has very little correlation with $F$ and if $h_i^2\approx 1$ then it indicates that $z_i$ has very strong correlation with $F$ and hence with the rest of the $z_j$ as well.
