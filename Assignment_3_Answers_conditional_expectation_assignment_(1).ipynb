# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

```python
r"""# Q2) Conditional Density of the Multivariate Gaussian Derivation

We want to show that the conditional distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$ is given by:
$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X|Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$

## Derivation:

1.  **Start with the definition of conditional probability density function:**
    $$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

2.  **Recall the joint PDF of a multivariate Gaussian $\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix}$:**
    $$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right)$$
    where $\mathbf{z} = \begin{bmatrix} \mathbf{x} \\ \mathbf{y} \end{bmatrix}$ and $\boldsymbol{\mu} = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix}$.

3.  **Recall the marginal PDF of $\mathbf{Y}$:**
    Since $\mathbf{Z}$ is multivariate Gaussian, its marginal distributions are also Gaussian. Thus, $\mathbf{Y} \sim \mathcal{N}(\boldsymbol{\mu}_Y, \boldsymbol{\Sigma}_{YY})$.
    $$f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)$$

4.  **Key properties for Gaussian distributions:**
    For a partitioned multivariate Gaussian, the quadratic form in the exponent of the joint PDF can be decomposed, and the determinant of the covariance matrix can be factored:
    *   **Determinant identity:** $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| |\boldsymbol{\Sigma}_{X|Y}|$
    *   **Quadratic form decomposition:** The exponent term $- \frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu})$ can be rewritten as:
        $$ - \frac{1}{2} \left[ (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right] $$
        where $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$.

5.  **Substitute these into the conditional PDF formula:**

    $$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}_{YY}| |\boldsymbol{\Sigma}_{X|Y}|}} \exp\left( -\frac{1}{2} [ (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) ] \right)}{\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)}$$

6.  **Simplify:**

    The term $\exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)$ cancels out from the numerator and denominator.

    The normalization constant simplifies to:
    $$ \frac{\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}_{YY}| |\boldsymbol{\Sigma}_{X|Y}|}}}{\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}} = \frac{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}_{YY}| |\boldsymbol{\Sigma}_{X|Y}|}} = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X|Y}|}} $$

    Combining the simplified terms, we obtain the conditional PDF:

    $$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X|Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

This shows that the conditional distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$ is also a Gaussian distribution with mean $\boldsymbol{\mu}^*$ and covariance matrix $\boldsymbol{\Sigma}_{X|Y}$."""
```

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



```python
import numpy as np
import math

# Given parameters
mu_X = 0
sigma_X_sq = 1
sigma_X = np.sqrt(sigma_X_sq)

mu_Q = 0
sigma_Q_sq = 4
sigma_Q = np.sqrt(sigma_Q_sq)

# X and Q are independent

print("--- Q3.1: Joint Distribution ---")

# Distribution of Y
# Y = X + Q
# Since X and Q are independent normal, Y is also normal.
mu_Y = mu_X + mu_Q
sigma_Y_sq = sigma_X_sq + sigma_Q_sq
sigma_Y = np.sqrt(sigma_Y_sq)

print(f"Y is normally distributed: Y ~ N({mu_Y}, {sigma_Y_sq})")
print(f"Verification: Y ~ N(0, 5) is correct.")

# Covariance(X, Y)
# Cov(X, Y) = Cov(X, X + Q)
# Cov(X, Y) = Cov(X, X) + Cov(X, Q)
# Since X and Q are independent, Cov(X, Q) = 0.
# Cov(X, Y) = Var(X)
cov_XY = sigma_X_sq
print(f"Cov(X, Y) = {cov_XY}")

print("\n--- Q3.2: Information Geometry ---")
print("X is not σ(Y)-measurable because Y = X + Q, where Q is a random variable with non-zero variance.")
print("This means X = Y - Q. Since Q is random, knowing Y does not uniquely determine X.")
print("For instance, if Y=5, X could be 1 (if Q=4) or X could be 4 (if Q=1), or any other combination.")
print("The power meter reading (Y) mixes the true temperature (X) with mechanical noise (Q),")
print("making it impossible to perfectly infer X from Y alone.")

print("\n--- Q3.3: Conditional Expectation (The Best Estimate) ---")

# Compute E[X | Y]
# E[X | Y = y] = mu_X + (Cov(X,Y) / Var(Y)) * (y - mu_Y)
conditional_coefficient = cov_XY / sigma_Y_sq
E_X_given_Y_str = f"{mu_X} + ({cov_XY}/{sigma_Y_sq}) * (y - {mu_Y})"
print(f"E[X | Y = y] = {E_X_given_Y_str}")
print(f"E[X | Y = y] = {conditional_coefficient:.4f} * y")
print(f"So, E[X | Y] = {conditional_coefficient:.4f}Y")

# Express as a "signal-to-total-variance" ratio
signal_to_total_variance_ratio = sigma_X_sq / (sigma_X_sq + sigma_Q_sq)
print(f"This is also expressed as (Var(X) / (Var(X) + Var(Q))) * Y = ({sigma_X_sq} / ({sigma_X_sq} + {sigma_Q_sq})) * Y = {signal_to_total_variance_ratio:.4f}Y")


print("\n--- Q3.4: Verification of the Residual ---")

# Prove E[Z] = 0 where Z = X - E[X | Y]
# E[Z] = E[X - E[X | Y]]
# By Law of Total Expectation, E[E[X | Y]] = E[X].
# E[Z] = E[X] - E[X] = 0
E_Z = mu_X - mu_X
print(f"E[Z] = E[X - E[X | Y]] = E[X] - E[E[X | Y]] = E[X] - E[X] = {E_Z}")
print(f"Since E[X] = {mu_X}, E[Z] = {E_Z}. (Proven)")

# Verify Z is uncorrelated with Y
# Cov(Z, Y) = E[ZY] - E[Z]E[Y]
# Since E[Z] = 0, Cov(Z, Y) = E[ZY]
# Z = X - (cov_XY / sigma_Y_sq)Y
# E[ZY] = E[(X - (cov_XY / sigma_Y_sq)Y)Y]
#        = E[XY - (cov_XY / sigma_Y_sq)Y^2]
#        = E[XY] - (cov_XY / sigma_Y_sq)E[Y^2]
# E[XY] = Cov(X, Y) + E[X]E[Y] = cov_XY + mu_X * mu_Y = 1 + 0*0 = 1
# E[Y^2] = Var(Y) + (E[Y])^2 = sigma_Y_sq + mu_Y^2 = 5 + 0^2 = 5
E_XY = cov_XY + mu_X * mu_Y
E_Y_sq = sigma_Y_sq + mu_Y**2

E_ZY = E_XY - (conditional_coefficient * E_Y_sq)
print(f"Cov(Z, Y) = E[ZY] - E[Z]E[Y] = E[ZY] - 0*{mu_Y} = E[ZY]")
print(f"E[XY] = Cov(X,Y) + E[X]E[Y] = {cov_XY} + {mu_X}*{mu_Y} = {E_XY}")
print(f"E[Y^2] = Var(Y) + E[Y]^2 = {sigma_Y_sq} + {mu_Y}^2 = {E_Y_sq}")
print(f"E[ZY] = E[XY] - (Cov(X,Y)/Var(Y)) * E[Y^2] = {E_XY} - ({conditional_coefficient} * {E_Y_sq}) = {E_ZY}")
print(f"Since E[ZY] = {E_ZY}, Z and Y are uncorrelated. (Verified - Orthogonality Principle)")

print("\n--- Q3.5: Information Theory: Differential Entropy ---")

# Compute h(X)
# For N(mu, sigma^2), h(X) = 0.5 * log2(2 * pi * e * sigma^2)
h_X = 0.5 * math.log2(2 * math.pi * math.e * sigma_X_sq)
print(f"Differential entropy h(X) = 0.5 * log2(2 * pi * e * Var(X)) = {h_X:.4f} bits")

# Compute h(X | Y)
# Variance of the conditional distribution (X | Y=y)
sigma_X_given_Y_sq = sigma_X_sq - (cov_XY**2 / sigma_Y_sq)
h_X_given_Y = 0.5 * math.log2(2 * math.pi * math.e * sigma_X_given_Y_sq)
print(f"Conditional variance Var(X | Y) = Var(X) - Cov(X,Y)^2 / Var(Y) = {sigma_X_sq} - ({cov_XY}^2 / {sigma_Y_sq}) = {sigma_X_given_Y_sq:.4f}")
print(f"Conditional differential entropy h(X | Y) = 0.5 * log2(2 * pi * e * Var(X | Y)) = {h_X_given_Y:.4f} bits")

print("\n--- Q3.6: Mutual Information ---")

# Calculate I(X; Y) = h(X) - h(X | Y)
I_X_Y = h_X - h_X_given_Y
print(f"Mutual Information I(X; Y) = h(X) - h(X | Y) = {I_X_Y:.4f} bits")
# Alternative calculation I(X;Y) = 0.5 * log2(Var(X) / Var(X|Y))
I_X_Y_alt = 0.5 * math.log2(sigma_X_sq / sigma_X_given_Y_sq)
print(f"Alternative I(X; Y) = 0.5 * log2(Var(X) / Var(X|Y)) = {I_X_Y_alt:.4f} bits (matches)")

# Percentage of uncertainty removed
percentage_removed = (I_X_Y / h_X) * 100
print(f"Percentage of uncertainty about temperature removed by observing power meter = {percentage_removed:.2f}%")

# How would this change if mechanical noise Q had a variance of 100?
sigma_Q_sq_new = 100
sigma_Y_sq_new = sigma_X_sq + sigma_Q_sq_new # 1 + 100 = 101
cov_XY_new = sigma_X_sq # still 1
sigma_X_given_Y_sq_new = sigma_X_sq - (cov_XY_new**2 / sigma_Y_sq_new) # 1 - (1^2 / 101) = 100/101
h_X_given_Y_new = 0.5 * math.log2(2 * math.pi * math.e * sigma_X_given_Y_sq_new)
I_X_Y_new = h_X - h_X_given_Y_new
percentage_removed_new = (I_X_Y_new / h_X) * 100

print(f"\nIf mechanical noise Q had a variance of {sigma_Q_sq_new}:")
print(f"New Conditional variance Var(X | Y) = {sigma_X_given_Y_sq_new:.4f}")
print(f"New Mutual Information I(X; Y) = {I_X_Y_new:.4f} bits")
print(f"New Percentage of uncertainty removed = {percentage_removed_new:.2f}%")
print("As the noise (Var(Q)) increases, the conditional variance of X given Y increases,")
print("leading to a decrease in mutual information and a smaller percentage of uncertainty removed.")
print("This means the power meter becomes less informative about the temperature when there is more mechanical noise.")

print("\n--- Q3.7: Diagnostic Interpretation ---")
print("The factory manager's claim: 'our best estimate of temperature should just be Y itself.'")
print("The manager's estimate is Y. Our calculated best estimate (conditional expectation) is E[X | Y] = 0.2Y.")

print("Why Y is a biased estimator of X (for a given observation y):")
print("If we observe Y=y, the manager's estimate is y. However, the true conditional mean of X is E[X | Y=y] = 0.2y.")
print("Unless y=0, these two are different. For example, if Y=5, the manager estimates X=5, but E[X|Y=5]=1.")
print("This means the manager's estimator overestimates X when Y > 0 and underestimates when Y < 0.")
print("So, for a specific observation Y=y, the manager's estimator 'y' is biased for X, as it includes the noise Q.")

print("Why E[X | Y] 'shrinks' the measurement toward the mean (zero):")
print("E[X | Y] = (Var(X) / (Var(X) + Var(Q))) * Y = 0.2Y.")
print("Since Var(X) < Var(X) + Var(Q), the coefficient (0.2) is less than 1.")
print("This means that the estimate E[X | Y] is always closer to the mean of X (which is 0) than the observed Y value itself, unless Y=0.")
print("For example, if Y=5, E[X|Y]=1. The observed Y is 5 units away from the mean 0, but the estimate for X is only 1 unit away from 0.")
print("This 'shrinkage' occurs because the observation Y contains noise Q. When we observe Y, we know some of that value is due to X and some to Q.")
print("The conditional expectation weighs the observation by the signal-to-noise ratio, effectively downplaying the contribution of the noise and pulling the estimate back towards the prior mean of X.")
print("This is a common phenomenon in Bayesian estimation and Wiener filtering, where estimates are pulled towards the prior mean if the measurement is noisy.")
```
