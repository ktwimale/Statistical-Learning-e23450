# Q12) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*

## Sample Answer

Here is the detailed worked-out solution for the CNC Spindle diagnostic problem.

---

**1. Joint Distribution of $(X, Y)$:**

Since $X$ and $Q$ are independent normal random variables, their sum $Y = X + Q$ is also normally distributed.
* **Mean:** $E[Y] = E[X] + E[Q] = 0 + 0 = 0$.
* **Variance:** Since $X \perp Q$, $\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$.
* **Covariance:** $\text{Cov}(X, Y) = E[XY] - E[X]E[Y] = E[X(X+Q)] = E[X^2] + E[XQ]$.
    * Since $E[X^2] = \text{Var}(X) = 1$ and $E[XQ] = E[X]E[Q] = 0$:
    * **$\text{Cov}(X, Y) = 1$**.

The joint distribution is:
$$\begin{pmatrix} X \\ Y \end{pmatrix} \sim \mathscr{N} \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & 1 \\ 1 & 5 \end{pmatrix} \right)$$

---

**2. Information Geometry:**

$X$ is not $\sigma(Y)$-measurable because there is no deterministic function $g$ such that $X = g(Y)$. For any observed $Y=y$, there are infinitely many pairs $(x, q)$ such that $x+q=y$.
**Practically:** A high power reading ($Y$) is ambiguous; it could be a "Thermal Fault" (high $X$) or simply a "Hard Material" (high $Q$). The sensor $Y$ cannot distinguish between these two physical states.

---

**3. Conditional Expectation (The Best Estimate):**

For bivariate normal variables, the best mean-square estimate is:
$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0)$$
$$\mathbf{E[X \mid Y] = 0.2 Y}$$
**Interpretation:** Because the noise ($Q$) is 4 times more volatile than the signal ($X$), we only "trust" 20% of the power fluctuation as being caused by temperature.

---

**4. Verification of the Residual:**

Let $\widehat{X} = 0.2 Y$. The residual is $Z = X - 0.2 Y$.
* **Mean:** $E[Z] = E[X] - 0.2 E[Y] = 0 - 0 = 0$.
* **Orthogonality:** $\text{Cov}(Z, Y) = E[ZY] = E[(X - 0.2Y)Y] = E[XY] - 0.2 E[Y^2]$.
    * $E[XY] = 1$ and $E[Y^2] = 5$.
    * $\text{Cov}(Z, Y) = 1 - 0.2(5) = 1 - 1 = 0$.
The error is uncorrelated with our observation, meaning we have extracted all possible information from $Y$.

---

**5. Differential Entropy:**

Recall $h(\mathscr{N}(\mu, \sigma^2)) = \frac{1}{2}\ln(2\pi e \sigma^2)$.
* **Marginal $h(X)$:** With $\sigma_X^2 = 1$:
$$h(X) = \frac{1}{2}\ln(2\pi e) \approx 1.4189 \text{ nats}$$
* **Conditional $h(X|Y)$:** The conditional variance is
$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1}{5} = 0.8.$$
$$h(X|Y) = \frac{1}{2}\ln(2\pi e \cdot 0.8) \approx 1.3073 \text{ nats}$$

---

**6. Mutual Information:**

$$I(X; Y) = h(X) - h(X|Y) = \frac{1}{2}\ln(1) - \frac{1}{2}\ln(0.8) = \frac{1}{2}\ln\left(\frac{1}{0.8}\right) = \frac{1}{2}\ln(1.25) \approx \mathbf{0.1116 \text{ nats}}$$
**Percentage Uncertainty Removed:** This is often measured by the reduction in variance. We went from a variance of $1.0$ to $0.8$ (a **20% reduction**). If the noise $Q$ increased to $\sigma_Q^2 = 100$, $I(X;Y)$ would approach 0, as the temperature signal would be "drowned out."

---

**7. Diagnostic Interpretation:**

The manager's suggestion to use $X = Y$ is a **Naive Estimator**.
* If we use $X = Y$, our error variance would be $E[(X-Y)^2] = E[Q^2] = 4$.
* By using $E[X|Y] = 0.2Y$, our error variance is only $0.8$.

**Shrinkage:** The factor $0.2$ is called a shrinkage estimator. It tells us that when a sensor is noisy, the "best" guess is to be conservative and pull the reading back toward the historical average (zero). If the power spikes, we don't immediately assume the tool is melting; we assume most of that spike is just the material being tough.
